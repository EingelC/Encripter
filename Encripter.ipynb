### 1 - Primero importamos las librerías:

```python
import json
from cryptography.fernet import Fernet
```

<a name='2'></a>
### 2 - Creamos nuestra llave:
* Debe modificar la ruta, en el ejemplo se muestra `D:/Fleet Ticket Automatization/`, aquí deberá poner donde planea guardar la ruta.
* Si modifica el nombre de `secret.key` deberá cambiarlo en el resto del código, le sugerimos no modificarlo.
* Este código solo debe ejecutarlo la primera vez. ¡Nunca deberá volver a ejecutar!
* No comparta con nadie su llave, será indispensable para mantener la seguridad del sistema.

```python
key = Fernet.generate_key()
with open("D:/Fleet Ticket Automatization/secret.key", "wb") as f:
    f.write(key)
print("Clave generada:", key)
```

<a name='3'></a>
### 3 - ¡Encriptemos TODO!
* Leemos la llave generada anteriormente.
* El segundo archivo a abrir serán sus credenciales, recuerde que deberá generarlas usted mismo o solicitarlas por una tarifa.
* Deberá guardar con mucho cuidado los archivos con termianción `.key` y `.enc`

```python
# Ruta de su llave privada
with open("D:/Fleet Ticket Automatization/secret.key", "rb") as f:
    key = f.read()
fernet = Fernet(key)

# Cargar credenciales
with open("D:/Fleet Ticket Automatization/jira.json", "rb") as f:
    data = f.read()

# Encriptar
encrypted = fernet.encrypt(data)

# Guardar cifrado
with open("D:/Fleet Ticket Automatization/jira_credentials.enc", "wb") as f:
    f.write(encrypted)

print("Archivo encriptado creado.")
```

Si pudo guardar con éxito su llave y sus credenciales encriptadas, ¡No tiene que volver a hacerlo nunca más!
